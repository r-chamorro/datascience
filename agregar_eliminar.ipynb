Como eliminar columnas de un DF

```python
df_books.drop("Year",axis=1).head(2)
```

```python
# Es posible eliminar las columnas para mostrar un conjunto de datos mas ordenados
# O adecuados a la necesidad, esto no afecta el DF original
df_books.drop(["Year","User Rating", "Price","Reviews","Genre"],axis=1).head(2)
```

```python
# Llamando al DF original, vemos como todo sigue igual sin tener solo un efecto visual
df_books.head(2)
```

```python
# Por otro lado, si agregas el parametro "inplace=True" esto se aplica al DF
df_books.drop(["Year","User Rating", "Price","Reviews","Genre"],axis=1, inplace=True)
```

```python
df_books.head(5)
# Esta ves podemos ver que el resultado si ha sido limpiado del DF.
```

```python
df_books.drop([0,3],axis=0).head(2)
# Esta vez borramos una fila en especifico, con este comando podriamos limpiar la data para hacerla
# Mas acorde a las necesidades del proyecto.
```

```python
# Para agregar columnas, recurrimos al comando Add columns.
df_books["Nueva_Columna"]= ((2024-df_books["Year"])*df_books["Price"])
df_books.head(4)
```

```python
df_books.drop(columns="Nueva_Columna", inplace=True)
```

```python
df_books.head(4)
```

```python
df_books.shape[0]
# Esto retorna una lista de 2 elementos, siendo el primero 0 la cantidad de Lineas y el segundo 1 la cantidad de columnas
# Nuestro DF contiene, al seleccionar 0 solo contamos las filas
```

```python
data = np.arange(0, df_books.shape[0])
data
# Creando un arreglo de 550 registros, numeros desde 0 hasta 549 estos pueden luego ser agregados
# Al DF que estamos trabajando
```

```python
df_books["RollingCounter"] = data
# Toda informacion que vayamos a ingresar debe tener la misma cantidad de filas que nuestro DF receptor
```
